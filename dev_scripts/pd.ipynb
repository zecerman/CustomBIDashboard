```python
# This Notebook is used for developer side EDA, it is not part of the final product

import pandas as pd
import numpy as np
```

```python
file_path = "TXT Files/MainData.txt"
```

```python
def read_to_df(file_path):
    df = pd.read_csv(
        file_path,
        sep=r"\|\|",
        engine="python",
        encoding="cp1252",
        dtype=str
    )

    # Clean quotes and whitespace
    df = df.apply(lambda col: col.str.strip().str.strip('"'))
    df.columns = df.columns.str.strip().str.strip('"')

    # Infer types
    df = df.apply(pd.to_numeric, errors="ignore")
    return df
```

```python
# Read single txt file into a df
MainData = read_to_df(file_path)
```

```python
df = MainData.copy()
df = df[df['CalendarID'] == 6]
df = df[df['StoreID'] > 0]
df = df.sort_values(by='StoreID')
df['StoreID'].head(200)
```
